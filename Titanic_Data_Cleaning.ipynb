```python
# Importing dependencies
import pandas as pd
```

```python
# Reading data from excel file
data = pd.read_excel("Week_04_HW4_Titanic_Data.xls",header=0)
```

```python
# Display top 5 rows to the feel of the data
data.head()
```

```python
# Displaying total no. of observations and attributes in the dataset
data.shape
```

```python
# Displaying null value counts for all the attributes
data.isnull().sum()
```

```python
# Dropped Cabin and Life Boat attributes as they have too many missing values
data = data.drop(['Cabin','Life Boat'],axis=1)
data.head()
```

```python
# Filling missing values of age with average value
data['Age'] = data['Age'].fillna(data['Age'].mean())
data.isnull().sum()
```

```python
# Rows with missing values are dropped
data = data.dropna()
data.isnull().sum()
```
